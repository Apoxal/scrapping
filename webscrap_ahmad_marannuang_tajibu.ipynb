```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
```

**coba 1 page**

```python
url = 'https://tv.lk21official.my/populer/'
session = requests.Session()
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = session.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    titles = soup.find_all('h1', class_='grid-title')


    movie_titles = []

    for title in titles:
        movie_title = title.find('a').text.strip()
        movie_titles.append(movie_title)


    for movie_title in movie_titles:
        print(f"Title: {movie_title}")

```

**coba untuk menampilkan deskripsi film**

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    titles = soup.find_all('h1', class_='grid-title')

    movie_data = []

    for title in titles:
        movie_title = title.find('a').text.strip()

        footer = title.find_next('footer')
        categories = footer.find('div', class_='grid-categories').find_all('a')

        categories_data = [category.text.strip() for category in categories]

        movie_data.append({
            'title': movie_title,
            'categories': categories_data
        })

    for movie in movie_data:
        print(f"Title: {movie['title']}")
        print("Categories:", ", ".join(movie['categories']))
        print()

else:
    print(f"Failed to retrieve page, status code: {response.status_code}")
```

**coba menjumlahkan film berdasarkan genre/kategori**

```python
url = 'https://tv.lk21official.my/populer/'
headers = {'User-Agent': 'Mozilla/5.0'}


response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    titles = soup.find_all('h1', class_='grid-title')

    category_count = defaultdict(int)
    movie_data = []

    for title in titles:
        movie_title = title.find('a').text.strip()
        footer = title.find_next('footer')
        categories = footer.find('div', class_='grid-categories').find_all('a')
        categories_data = [category.text.strip() for category in categories]

        movie_data.append({'title': movie_title, 'categories': categories_data})
        for category in categories_data:
            category_count[category] += 1

    for movie in movie_data:
        print(f"Title: {movie['title']}, Categories: {', '.join(movie['categories'])}")

    print("\nFrekuensi Kategori:")
    for category, count in category_count.items():
        print(f"Genre {category}: {count}")

else:
    print(f"Failed to retrieve page, status code: {response.status_code}")
```

**coba crawling 2: menjumlahkan film berdasarkan kategori**

```python
popular_url = 'https://tv.lk21official.my/populer/page/{}'
most_commented_url = 'https://tv.lk21official.my/most-commented/page/{}'
headers = {'User-Agent': 'Mozilla/5.0'}

movie_data = []
category_count = defaultdict(int)

def scrape_movies(url):
    """Fungsi untuk mengambil data film dari halaman tertentu."""
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        titles = soup.find_all('h1', class_='grid-title')

        for title in titles:
            movie_title = title.find('a').text.strip()
            footer = title.find_next('footer')

            categories = footer.find('div', class_='grid-categories').find_all('a')
            categories_data = [category.text.strip() for category in categories]

            movie_data.append({'title': movie_title, 'categories': categories_data})


            for category in categories_data:
                category_count[category] += 1
    else:
        print(f"Failed to retrieve page, status code: {response.status_code}")


for page in range(1, 6):
    url = popular_url.format(page)
    scrape_movies(url)


for page in range(1, 6):  #
    url = most_commented_url.format(page)
    scrape_movies(url)


for movie in movie_data:
    print(f"Title: {movie['title']}, Categories: {', '.join(movie['categories'])}")

print("\nFrekuensi Kategori:")
for category, count in category_count.items():
    print(f"Genre {category}: {count}")
```
